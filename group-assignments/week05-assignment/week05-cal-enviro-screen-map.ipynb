### The following code maps Vulnerability Scores provided by the CalEnviro Screen 3.0 survey.

##### Read in census population data, show shape and first five rows.

```python
import pandas as pd
import matplotlib.pyplot as plt
cal_data = pd.read_csv('calenviroscreen-3.0-results.csv')
print(cal_data.shape)
cal_data.head()
```

##### Let's look at all columns in the data.

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Display a random row in the dataset.
cal_data.sample()
```

##### Now examine the data types.

```python
cal_data.info()
```

##### Get rid of null columns and reinspect a sample of the data. No columns were null, therefore none were dropped.

```python
cal_data.columns[cal_data.isna().all()].tolist()
cal_data = cal_data.dropna(axis=1,how="all")

cal_data.info()
cal_data.sample()
```

##### Identify columns to keep.

```python
columns_to_keep = ['Census Tract',
                   'Total Population',
                   'CES 3.0 Score',
                   'SB 535 Disadvantaged Community',
                  ]
```

##### Create new dataset with all necessary columns.

```python
cal_data_final = cal_data[columns_to_keep]
cal_data_final.head()
```

##### Rename columns.

```python
columns = list(cal_data_final) # this is the same as df.columns.to_list()
cal_data_final.columns = ['census_tract',
'total_pop',
'score',
'sb_535_disadvantaged_community']
# See new columns applide to dataset.
cal_data_final.sample(5)
```

##### Need to add a FIPS column with adjusted census-tract ID numbers.

```python
cal_data_final['FIPS'] ='0' + cal_data_final['census_tract'].astype(str)
cal_data_final.sample(5)
```

##### Import geopandas library for mapping and read in census tract data.

```python
import geopandas as gpd
# read in a geojson file downloaded from the LA Times
tracts = gpd.read_file('census_tracts_2010.geojson')
tracts.head()
```

##### Exclude all columns from dataset except FIPS and geometry from census tract geography data.

```python
tracts = tracts[['GEOID10','geometry']]
tracts.rename(columns = {'GEOID10':'FIPS'}, inplace = True)
tracts.head()
```

##### Join the census tract data and CalEnviro Screen data based on FIPS.

```python
tracts_calenviro = tracts.merge(cal_data_final,on="FIPS")
tracts_calenviro.head()
```

##### Map CES 3.0 Scores across census tracts.

```python
print("CalEnviro Screen in LA County Census Tracts")
tracts_calenviro.plot(figsize=(12,10),
                 column='score',
                 legend=True, 
                 scheme='NaturalBreaks')
```
