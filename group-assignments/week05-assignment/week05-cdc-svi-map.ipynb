### The following code maps the CDC's Social Vulnerability Index across LA census tracts.

##### Read in census population data, show shape and first five rows.

```python
import pandas as pd
import matplotlib.pyplot as plt
svi_data = pd.read_csv('cdc-svi-data.csv')
print(svi_data.shape)
svi_data.head()
```

##### Let's look at all columns in the data.

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# Display a random row in the dataset.
svi_data.sample()
```

##### Now examine the data types.

```python
svi_data.info()
```

##### Get rid of null columns and reinspect a sample of the data. No columns were null, therefore none were dropped.

```python
svi_data.columns[svi_data.isna().all()].tolist()
svi_data = svi_data.dropna(axis=1,how="all")

svi_data.info()
```

##### Identify columns to keep.

```python
columns_to_keep = ['FIPS',
                   'RPL_THEME1',
                   'RPL_THEME2',
                   'RPL_THEME3',
                   'RPL_THEME4',
                   'RPL_THEMES'
                  ]
```

##### Create new dataset with all necessary columns.

```python
svi_data_final = svi_data[columns_to_keep]
svi_data_final.head()
```

##### Rename columns.

```python
columns = list(svi_data_final) # this is the same as df.columns.to_list()
svi_data_final.columns = ['FIPS',
'socio_econ_v_score',
'household_char_v_score',
'racial_ethnic_v_score',
'housing_transportation_v_score',
'final_v_score']
# See new columns applide to dataset.
svi_data_final.sample(5)
```

##### Import geopandas library for mapping and read in census tract data.

```python
import geopandas as gpd
# read in a geojson file downloaded from the LA Times
tracts = gpd.read_file('census_tracts_2010.geojson')
tracts.head()
```

##### Exclude all columns from dataset except FIPS and geometry from census tract geography data.

```python
tracts = tracts[['GEOID10','geometry']]
tracts.rename(columns = {'GEOID10':'FIPS'}, inplace = True)
tracts.head()
```

##### Join the census tract data and SVI data based on FIPS.

```python
#svi_data_final['FIPS'] = svi_data_final['FIPS'].astype(int)
#tracts['FIPS'] = tracts['FIPS'].astype(int)

tracts_svi = tracts.merge(svi_data_final,on="FIPS")
tracts_svi.head()
```

```python
print("Social Vulnerability Scores in LA Census Tracts")
tracts_svi.plot(figsize=(12,10),
                 column='final_v_score',
                 legend=True, 
                 scheme='NaturalBreaks')
```
