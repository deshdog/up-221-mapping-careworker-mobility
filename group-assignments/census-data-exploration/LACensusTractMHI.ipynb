```python
import pandas as pd
```

```python
import geopandas as gpd
```

```python
imap = gpd.read_file('data/Median_Income_and_AMI_(census_tract).shp')
```

```python
imap.head()
```

```python
imap.plot()
```

```python
ax = imap.boundary.plot(edgecolor = 'black', linewidth = 0.0, figsize = (30,20))
imap.plot(ax = ax, column = 'med_hh_inc', cmap = 'YlGn', legend = True,
          legend_kwds = {'shrink':0.3, 'orientation' : 'horizontal', 'label' : 'Median Household Income ($)'})
ax.set_title('Los Angeles Median Household Income (by Census Tract)', size = 20, weight = 'bold')
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
for edge in ['right', 'left', 'bottom', 'top']:
    ax.spines[edge].set_visible(False)
```
