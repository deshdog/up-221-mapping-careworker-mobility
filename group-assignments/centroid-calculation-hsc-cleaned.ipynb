Write a function to caluclate the haversine distance between to lat-lon points. Found this on GIS StackExchange.

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 3956 # Radius of earth in miles. Use 6371 for kilometers
    return c * r
    
```

Let's select the Junipero Serra Branch Library as the location nearest to the Historic South Central neighborhood.

```python
lon1 = -118.274350
lat1 = 34.001730

# New list for distances
haversine_distances = []

for i,j in south_central_centroids.iterrows():
    coord = j
    lon2 = coord['lon']
    lat2 = coord['lat']
    haversine_var = haversine(lon1, lat1, lon2, lat2)
    haversine_distances.append(haversine_var)
    
print("List of distances from each tract centroid to the Junipero Serra Branch Library:")
print()
print(haversine_distances)
```

Now write a function that finds the average value of a list. Use it to calculate the average distance from census tract centroids to the nearest library.

```python
def average(lst):
    return sum(lst) / len(lst)
avg_dist_to_library = average(haversine_distances) 
print("average distance to nearest public library", avg_dist_to_library, "miles")
```
